```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
# Get all CSV files from new_scraped directory
csv_dir = Path('.')
csv_files = sorted([f for f in csv_dir.glob('*.csv') if f.name != 'All_Roles.csv'])
print(f"Found {len(csv_files)} CSV files")
for f in csv_files:
    print(f"  - {f.name}")
```

```python
# Function to create visualization for each CSV
def visualize_skills(csv_path, top_n=20):
    # Extract job title from filename (remove .csv extension)
    job_title = csv_path.stem.replace('_', ' ')
    
    # Read CSV
    df = pd.read_csv(csv_path)
    
    # Take top N skills
    df_top = df.head(top_n).copy()
    
    # Create figure with 2 subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
    
    # Bar chart - Count
    sns.barplot(data=df_top, y='Skill', x='Count', ax=ax1, palette='viridis', hue='Skill', legend=False)
    ax1.set_title(f'{job_title} - Top {top_n} Skills by Count', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Count', fontsize=12)
    ax1.set_ylabel('Skill', fontsize=12)
    
    # Bar chart - Percentage
    # Remove % sign and convert to float
    df_top['Percentage_val'] = df_top['Percentage'].str.rstrip('%').astype(float)
    sns.barplot(data=df_top, y='Skill', x='Percentage_val', ax=ax2, palette='mako', hue='Skill', legend=False)
    ax2.set_title(f'{job_title} - Top {top_n} Skills by Percentage', fontsize=14, fontweight='bold')
    ax2.set_xlabel('Percentage (%)', fontsize=12)
    ax2.set_ylabel('')
    
    plt.tight_layout()
    plt.show()
    
    # Print summary stats
    print(f"\n{job_title} Summary:")
    print(f"  Total skills: {len(df)}")
    print(f"  Top skill: {df.iloc[0]['Skill']} ({df.iloc[0]['Percentage']})")
    print(f"  Skills above 50%: {len(df[df['Percentage'].str.rstrip('%').astype(float) > 50])}")
    print("-" * 80)
```

# Comprehensive Cross-Role Analysis

```python
# Load all CSVs and aggregate data
all_data = {}
for csv_file in csv_files:
    role_name = csv_file.stem.replace('_', ' ')
    df = pd.read_csv(csv_file)
    all_data[role_name] = df

print(f"Loaded {len(all_data)} role datasets")
```

## 1. Top Skills Across All Roles (Universal Skills)

```python
# Aggregate skills across all roles
from collections import Counter

skill_counts = Counter()
for role_name, df in all_data.items():
    for skill in df['Skill']:
        skill_counts[skill] += 1

# Get top 20 most common skills across roles
top_universal_skills = pd.DataFrame(
    skill_counts.most_common(20),
    columns=['Skill', 'Number of Roles']
)

# Visualize
plt.figure(figsize=(12, 8))
sns.barplot(data=top_universal_skills, x='Number of Roles', y='Skill', palette='viridis', hue='Skill', legend=False)
plt.title('Most Common Skills Across All Job Roles', fontsize=16, fontweight='bold')
plt.xlabel('Number of Roles Requiring This Skill', fontsize=12)
plt.ylabel('Skill', fontsize=12)
plt.tight_layout()
plt.show()

print("\nTop 10 Universal Skills:")
for idx, row in top_universal_skills.head(10).iterrows():
    print(f"  {row['Skill']}: appears in {row['Number of Roles']} roles")
```

## 2. Skill Heatmap - Which Skills for Which Roles

```python
# Create a matrix of top skills vs roles
# Get top 15 most common skills
top_skills = [skill for skill, _ in skill_counts.most_common(15)]

# Create matrix
skill_role_matrix = pd.DataFrame(index=top_skills, columns=sorted(all_data.keys()))

for role_name, df in all_data.items():
    for skill in top_skills:
        if skill in df['Skill'].values:
            # Get percentage for this skill in this role
            pct = df[df['Skill'] == skill]['Percentage'].values[0]
            skill_role_matrix.loc[skill, role_name] = float(pct.rstrip('%'))
        else:
            skill_role_matrix.loc[skill, role_name] = 0

# Convert to numeric
skill_role_matrix = skill_role_matrix.astype(float)

# Plot heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(skill_role_matrix, annot=True, fmt='.1f', cmap='YlOrRd', cbar_kws={'label': 'Percentage (%)'})
plt.title('Skill Requirements Heatmap Across Job Roles', fontsize=16, fontweight='bold')
plt.xlabel('Job Role', fontsize=12)
plt.ylabel('Skill', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 3. Role Similarity Analysis

```python
# Calculate Jaccard similarity between roles based on their top 10 skills
from itertools import combinations

def jaccard_similarity(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union if union > 0 else 0

# Get top 10 skills for each role
role_top_skills = {}
for role_name, df in all_data.items():
    role_top_skills[role_name] = set(df.head(10)['Skill'].values)

# Calculate similarity matrix
roles = sorted(all_data.keys())
similarity_matrix = pd.DataFrame(index=roles, columns=roles)

for role1, role2 in combinations(roles, 2):
    sim = jaccard_similarity(role_top_skills[role1], role_top_skills[role2])
    similarity_matrix.loc[role1, role2] = sim
    similarity_matrix.loc[role2, role1] = sim

# Fill diagonal with 1.0
for role in roles:
    similarity_matrix.loc[role, role] = 1.0

# Convert to numeric
similarity_matrix = similarity_matrix.astype(float)

# Plot heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(similarity_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0.5, 
            vmin=0, vmax=1, cbar_kws={'label': 'Jaccard Similarity'})
plt.title('Role Similarity Based on Top 10 Skills', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

## 4. Most Similar and Distinct Role Pairs

```python
# Find most similar role pairs
similarities = []
for role1, role2 in combinations(roles, 2):
    sim = similarity_matrix.loc[role1, role2]
    similarities.append((role1, role2, sim))

# Sort by similarity
similarities.sort(key=lambda x: x[2], reverse=True)

print("Top 10 Most Similar Role Pairs:")
print("=" * 80)
for role1, role2, sim in similarities[:10]:
    print(f"{role1}  {role2}: {sim:.2%} similarity")

print("\n\nTop 10 Most Distinct Role Pairs:")
print("=" * 80)
for role1, role2, sim in similarities[-10:]:
    print(f"{role1}  {role2}: {sim:.2%} similarity")
```

## 5. Role-Specific vs Universal Skills

```python
# Identify role-specific skills (appear in only 1-2 roles)
role_specific_skills = {skill: count for skill, count in skill_counts.items() if count <= 2}
universal_skills = {skill: count for skill, count in skill_counts.items() if count >= 10}

print(f"Universal Skills (appear in 10+ roles): {len(universal_skills)}")
print(f"Role-Specific Skills (appear in 1-2 roles): {len(role_specific_skills)}")
print(f"\nUniversal Skills:")
for skill, count in sorted(universal_skills.items(), key=lambda x: x[1], reverse=True):
    print(f"  {skill}: {count} roles")

# Show role-specific skills for each role
print("\n" + "=" * 80)
print("Role-Specific Skills (unique or nearly unique):")
print("=" * 80)
for role_name, df in sorted(all_data.items()):
    specific = [skill for skill in df.head(10)['Skill'] if skill in role_specific_skills]
    if specific:
        print(f"\n{role_name}:")
        for skill in specific:
            print(f"  - {skill}")
```

## 6. Skill Demand Distribution

```python
# Analyze distribution of high-demand skills (>50%)
high_demand_summary = {}

for role_name, df in all_data.items():
    df['Percentage_val'] = df['Percentage'].str.rstrip('%').astype(float)
    high_demand = df[df['Percentage_val'] > 50]
    high_demand_summary[role_name] = len(high_demand)

# Sort by number of high-demand skills
sorted_summary = sorted(high_demand_summary.items(), key=lambda x: x[1], reverse=True)

# Visualize
plt.figure(figsize=(12, 8))
roles_sorted = [role for role, _ in sorted_summary]
counts = [count for _, count in sorted_summary]

sns.barplot(x=counts, y=roles_sorted, palette='rocket', hue=roles_sorted, legend=False)
plt.title('Number of High-Demand Skills (>50%) per Role', fontsize=16, fontweight='bold')
plt.xlabel('Number of Skills with >50% Demand', fontsize=12)
plt.ylabel('Job Role', fontsize=12)
plt.tight_layout()
plt.show()

print("\nRoles with Most High-Demand Skills:")
for role, count in sorted_summary[:5]:
    print(f"  {role}: {count} skills")
```

## 7. Summary Statistics

```python
# Overall statistics
total_unique_skills = len(skill_counts)
avg_skills_per_role = sum(len(df) for df in all_data.values()) / len(all_data)

print("=" * 80)
print("OVERALL SUMMARY")
print("=" * 80)
print(f"\nTotal Job Roles Analyzed: {len(all_data)}")
print(f"Total Unique Skills: {total_unique_skills}")
print(f"Average Skills per Role: {avg_skills_per_role:.1f}")
print(f"\nSkill Distribution:")
print(f"  - Universal (10+ roles): {len(universal_skills)}")
print(f"  - Common (5-9 roles): {len([s for s, c in skill_counts.items() if 5 <= c <= 9])}")
print(f"  - Moderate (3-4 roles): {len([s for s, c in skill_counts.items() if 3 <= c <= 4])}")
print(f"  - Role-specific (1-2 roles): {len(role_specific_skills)}")

print(f"\n\nKey Insights:")
print(f"1. Python is the most universal skill, appearing in {skill_counts['Python']} roles")
print(f"2. Most similar roles: {similarities[0][0]}  {similarities[0][1]} ({similarities[0][2]:.2%})")
print(f"3. Most distinct roles: {similarities[-1][0]}  {similarities[-1][1]} ({similarities[-1][2]:.2%})")
print(f"4. Role with most high-demand skills: {sorted_summary[0][0]} ({sorted_summary[0][1]} skills)")
print("=" * 80)
```

## 8. Career Transition Roadmap - Which Roles Are Easiest to Switch To?

```python
# For each role, find the 3 easiest roles to transition to
print("CAREER TRANSITION RECOMMENDATIONS")
print("=" * 100)
print("Format: [Current Role]  [Target Role] (Skill Overlap %)\n")

for role in sorted(roles):
    # Get similarities for this role
    role_sims = [(other, similarity_matrix.loc[role, other]) 
                 for other in roles if other != role]
    role_sims.sort(key=lambda x: x[1], reverse=True)
    
    print(f"\n{role}:")
    print(f"  Easiest transitions:")
    for target, sim in role_sims[:3]:
        print(f"     {target} ({sim:.0%} skill overlap)")
    
    # Calculate skills gap
    current_skills = set(all_data[role].head(10)['Skill'].values)
    target_skills = set(all_data[role_sims[0][0]].head(10)['Skill'].values)
    missing_skills = target_skills - current_skills
    if missing_skills:
        print(f"  Skills to learn for {role_sims[0][0]}: {', '.join(list(missing_skills)[:3])}")
```

## 9. Role Differentiators - What Makes Each Role Unique?

```python
# Find unique/rare skills that differentiate each role
print("ROLE DIFFERENTIATORS - Key Skills That Set Each Role Apart")
print("=" * 100)

for role_name in sorted(all_data.keys()):
    df = all_data[role_name]
    df['Percentage_val'] = df['Percentage'].str.rstrip('%').astype(float)
    
    # Get skills that appear in this role but not in many others
    unique_skills = []
    for skill in df.head(15)['Skill']:
        if skill_counts[skill] <= 3:  # Rare skill
            pct = df[df['Skill'] == skill]['Percentage_val'].values[0]
            unique_skills.append((skill, pct))
    
    if unique_skills:
        print(f"\n{role_name}:")
        unique_skills.sort(key=lambda x: x[1], reverse=True)
        for skill, pct in unique_skills[:5]:
            roles_with_skill = [r for r in all_data.keys() if skill in all_data[r]['Skill'].values]
            print(f"   {skill} ({pct:.1f}%) - Found in {len(roles_with_skill)} role(s)")
    else:
        # If no unique skills, show top 3 highest % skills
        top3 = df.head(3)
        print(f"\n{role_name} (mainstream role, focus on mastery):")
        for _, row in top3.iterrows():
            print(f"   {row['Skill']} ({row['Percentage']})")
```

## 10. Essential vs Optional Skills - Learning Priority

```python
# Classify skills by demand level for each role
print("LEARNING PRIORITY BY ROLE - Focus Your Effort")
print("=" * 100)
print("Legend:  Must-have (>70%) |  Important (50-70%) |  Nice-to-have (<50%)\n")

for role_name in sorted(all_data.keys()):
    df = all_data[role_name]
    df['Percentage_val'] = df['Percentage'].str.rstrip('%').astype(float)
    
    must_have = df[df['Percentage_val'] > 70].head(10)
    important = df[(df['Percentage_val'] >= 50) & (df['Percentage_val'] <= 70)].head(5)
    
    print(f"\n{role_name}:")
    
    if len(must_have) > 0:
        print(f"   Must-have skills:")
        for _, row in must_have.iterrows():
            print(f"     {row['Skill']} ({row['Percentage']})")
    
    if len(important) > 0:
        print(f"   Important skills:")
        for _, row in important.iterrows():
            print(f"     {row['Skill']} ({row['Percentage']})")
    
    # Count nice-to-haves
    nice_to_have_count = len(df[df['Percentage_val'] < 50])
    if nice_to_have_count > 0:
        print(f"   Nice-to-have: {nice_to_have_count} additional skills")
```

## 11. Skills Investment ROI - Which Skills Open Most Doors?

```python
# Calculate ROI for skills - how many roles + average demand %
print("SKILLS INVESTMENT ROI - Best Skills to Learn")
print("=" * 100)
print("Ranking based on: (Number of Roles)  (Average Demand %)\n")

skill_roi = {}
for skill, role_count in skill_counts.items():
    # Calculate average demand % across all roles that need this skill
    total_pct = 0
    count = 0
    for role_name, df in all_data.items():
        if skill in df['Skill'].values:
            pct = df[df['Skill'] == skill]['Percentage'].values[0]
            total_pct += float(pct.rstrip('%'))
            count += 1
    
    avg_pct = total_pct / count if count > 0 else 0
    roi_score = role_count * avg_pct
    skill_roi[skill] = {
        'roles': role_count,
        'avg_demand': avg_pct,
        'roi_score': roi_score
    }

# Sort by ROI score
sorted_roi = sorted(skill_roi.items(), key=lambda x: x[1]['roi_score'], reverse=True)

print("TOP 20 HIGHEST ROI SKILLS:\n")
for i, (skill, data) in enumerate(sorted_roi[:20], 1):
    print(f"{i:2d}. {skill:30s} | {data['roles']:2d} roles | {data['avg_demand']:5.1f}% avg demand | ROI: {data['roi_score']:6.1f}")

print("\n" + "=" * 100)
print("INTERPRETATION:")
print("   High ROI = Learn this skill to maximize career opportunities")
print("   Skills like Python, Machine Learning have high ROI (many roles + high demand)")
print("   Niche skills (NLP, Deep Learning) have lower ROI but essential for specific roles")
```

## 12. Competitive Analysis - How Crowded Is Each Role?

```python
# Analyze role competition based on skill requirements
print("ROLE COMPETITION ANALYSIS")
print("=" * 100)
print("Strategy: Niche roles = less competition, Mainstream roles = more competition\n")

role_analysis = []
for role_name, df in all_data.items():
    df['Percentage_val'] = df['Percentage'].str.rstrip('%').astype(float)
    
    # Metrics for competition
    total_skills = len(df)
    must_have_skills = len(df[df['Percentage_val'] > 70])
    top10_skills = set(df.head(10)['Skill'].values)
    
    # Check if skills are common across other roles
    common_skill_count = sum(1 for skill in top10_skills if skill_counts[skill] >= 10)
    
    # Competition score: lower = less competitive (niche)
    competition_score = (common_skill_count / 10) * 100  # % of common skills in top 10
    
    role_analysis.append({
        'role': role_name,
        'competition': competition_score,
        'total_skills': total_skills,
        'must_have': must_have_skills,
        'category': 'NICHE' if competition_score < 50 else 'COMPETITIVE'
    })

# Sort by competition
role_analysis.sort(key=lambda x: x['competition'])

print("NICHE ROLES (Less Competition, Specialized Skills):")
print("-" * 100)
for data in role_analysis[:8]:
    if data['category'] == 'NICHE':
        print(f"   {data['role']:30s} | Competition: {data['competition']:5.1f}% | {data['must_have']} must-have skills")

print("\n\nCOMPETITIVE ROLES (High Competition, Common Skills):")
print("-" * 100)
for data in role_analysis:
    if data['category'] == 'COMPETITIVE':
        print(f"   {data['role']:30s} | Competition: {data['competition']:5.1f}% | {data['must_have']} must-have skills")

print("\n" + "=" * 100)
print("STRATEGY:")
print("   Niche roles: Specialize deeply, less competition, harder to enter")
print("   Competitive roles: Need strong fundamentals + portfolio, more opportunities")
```

## 13. Actionable Career Playbook - 3 Strategies

```python
print("=" * 100)
print("ACTIONABLE CAREER PLAYBOOK")
print("=" * 100)

# Strategy 1: Generalist Path (maximize opportunities)
print("\n STRATEGY 1: GENERALIST PATH (Maximize Job Opportunities)")
print("-" * 100)
print("Target: Maximum role flexibility with universal skills\n")
top_universal = sorted_roi[:5]
print("Learn these 5 skills in order:")
for i, (skill, data) in enumerate(top_universal, 1):
    print(f"  {i}. {skill} - Opens {data['roles']} roles ({data['avg_demand']:.1f}% avg demand)")
print(f"\nRoles you'll qualify for: {len([r for r in all_data.keys() if all(skill in all_data[r]['Skill'].values for skill, _ in top_universal[:3])])} roles with top 3 skills")

# Strategy 2: Specialist Path (niche expertise)
print("\n\n STRATEGY 2: SPECIALIST PATH (Niche Expertise)")
print("-" * 100)
niche_roles = [r for r in role_analysis if r['category'] == 'NICHE'][:3]
print("Focus on these high-value niche roles:\n")
for data in niche_roles:
    df = all_data[data['role']]
    df['Percentage_val'] = df['Percentage'].str.rstrip('%').astype(float)
    must_haves = df[df['Percentage_val'] > 70].head(5)
    print(f"{data['role']}:")
    print(f"  Master these skills:")
    for _, row in must_haves.iterrows():
        print(f"     {row['Skill']} ({row['Percentage']})")
    print()

# Strategy 3: Transition Path (career pivot)
print("\n STRATEGY 3: SMART TRANSITION PATH (Career Pivot)")
print("-" * 100)
print("Recommended transition sequences (high skill overlap):\n")

# Find best 3-step career paths
best_paths = []
for role1 in roles[:10]:  # Limit for readability
    for role2 in roles:
        if role1 != role2:
            sim12 = similarity_matrix.loc[role1, role2]
            if sim12 > 0.4:  # Good overlap
                for role3 in roles:
                    if role3 not in [role1, role2]:
                        sim23 = similarity_matrix.loc[role2, role3]
                        if sim23 > 0.4:
                            avg_sim = (sim12 + sim23) / 2
                            best_paths.append((role1, role2, role3, avg_sim))

best_paths.sort(key=lambda x: x[3], reverse=True)
print("Top 5 smooth career progressions:")
for i, (r1, r2, r3, sim) in enumerate(best_paths[:5], 1):
    print(f"{i}. {r1}  {r2}  {r3} (avg {sim:.0%} overlap)")

print("\n" + "=" * 100)
print("CHOOSE YOUR STRATEGY BASED ON:")
print("   Generalist: Want flexibility, entry-level, or career change")
print("   Specialist: Deep expertise, higher pay ceiling, specific industry")
print("   Transition: Currently employed, want smooth career pivot")
print("=" * 100)
```
